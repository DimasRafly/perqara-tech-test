## Data Aggregation

```python
orders.info()
```

Data aggregation is to create a summary of the data that shows how many orders were placed during each time period. The code below aggregates data in daily timeframe.

```python
# Aggregatea data by date
daily_orders = orders.groupby(orders['order_purchase_timestamp'].dt.date).size()
daily_orders = daily_orders.rename('order_count').reset_index()
daily_orders['order_purchase_timestamp'] = pd.to_datetime(daily_orders['order_purchase_timestamp'])
daily_orders.set_index('order_purchase_timestamp', inplace=True)
```

## Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

The purpose of seasonal decomposition is to identifying the trend, seasonality, and residual components.

```python
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(daily_orders['order_count'])
plt.title('Daily Order Counts Over Time')
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_orders['order_count'], model='additive', period=30)
decomposition.plot()
plt.show()
```

- There seems to be a **general upward trend** in the number of orders over time and trend line is relatively smooth, **indicating a consistent upward direction**.
- The seasonal component **shows a clear periodic pattern**, suggesting that there are **recurring seasonal factors influencing** the number of orders.
- There are a **few outliers in the residuals**, which could be due to unusual events or anomalies in the data, especially around the end of 2017.

## Stationarity Check

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(daily_orders['order_count'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

The ADF **test results indicate that the time series is stationary**, meaning it **does not have a unit root**. So **differencing is not required** based on this situation.

---

The data can be define as Non-Stationary if the ADF test score **near zero or positive** and p-value is greater than the significance level (e.g., 0.05)

# **RFM Clustering**

To segment customers based on their purchasing behavior using the RFM (Recency, Frequency, Monetary) model. This approach can help identify valuable customer groups and tailor marketing strategies accordingly.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from datetime import datetime
```

## Load Data

```python
# Load Data
def load_data(orders_file, order_items_file):
    # Read the data path
    orders = pd.read_csv(orders_file)
    order_items = pd.read_csv(order_items_file)

    return orders, order_items
```

## Prepare RFM & Cluster Data

```python
# Prepare the RFM
def prepare_rfm_data(orders, order_items):
    #  Merge Dataset
    df = orders.merge(order_items, on='order_id')
    # Create new features
    df['total_purchase'] = df['price'] + df['freight_value']
    # Using Latest Date
    latest_date = pd.to_datetime(df['order_purchase_timestamp']).max()

    # Calculate RFM
    rfm = df.groupby('customer_id').agg({
        'order_purchase_timestamp': lambda x: (latest_date - pd.to_datetime(x.max())).days,
        'order_id': 'count',
        'total_purchase': 'sum'
    })

    # Assigning RFM
    rfm.columns = ['Recency', 'Frequency', 'Monetary']
    return rfm
```

## Function to Perform Clustering

```python
# Clustering Function
def perform_clustering(rfm, n_clusters=5):
    # Normalize the data
    scaler = StandardScaler()
    rfm_normalized = scaler.fit_transform(rfm)

    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(rfm_normalized)

    # Assigning cluster labels on RFM
    rfm = rfm.reset_index()
    rfm['Cluster'] = cluster_labels

    return rfm, kmeans
```

## Analyze Cluster Function

```python
# Summarize Cluster
def analyze_clusters(rfm):
    # Data Aggregation
    cluster_summary = rfm.groupby('Cluster').agg({
        'Recency': 'mean',
        'Frequency': 'mean',
        'Monetary': 'mean',
        'customer_id': 'count'
    }).reset_index()

    # Assigning New Columns
    cluster_summary.columns = ['Cluster', 'Avg_Recency', 'Avg_Frequency', 'Avg_Monetary', 'Count']
    return cluster_summary
```

## Main Function

```python
# Define number of cluster
def main(orders_file, order_items_file, n_clusters=5, n_init=10):
    # Load Data
    orders, order_items = load_data(orders_file, order_items_file)
    # Prepare RFM
    rfm = prepare_rfm_data(orders, order_items)
    # Perform Clustering
    rfm_clustered, kmeans_model = perform_clustering(rfm, n_clusters)
    # Analyze Cluster
    cluster_summary = analyze_clusters(rfm_clustered)
    # Return all arguments
    return rfm_clustered, cluster_summary, kmeans_model
```

# **Geoanalysis**

## Load Data

```python
# Load data
def load_data(geolocation_file, customers_file):
    geo_data = pd.read_csv(geolocation_file)
    customers = pd.read_csv(customers_file)

    # Merge customer data with geolocation data
    customer_locations = customers.merge(geo_data, left_on='customer_zip_code_prefix',
                                         right_on='geolocation_zip_code_prefix', how='left')
    return customer_locations
```

## Data Viz

```python
# Visualize the langitude and latitude
def visualize_customer_distribution(customer_locations):
    plt.figure(figsize=(12, 8))
    plt.scatter(customer_locations['geolocation_lng'], customer_locations['geolocation_lat'],
                alpha=0.1, s=1)
    plt.title('Customer Distribution in Brazil')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()

# Visualize the Number of Customer
def state_level_analysis(customer_locations):
    state_counts = customer_locations['geolocation_state'].value_counts()

    plt.figure(figsize=(12, 6))
    state_counts.plot(kind='bar')
    plt.title('Number of Customers by State')
    plt.xlabel('State')
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return state_counts
```

## Main Function

```python
# Assigning Previous Function
def main(geolocation_file, customers_file):
    customer_locations = load_data(geolocation_file, customers_file)

    # Top 5 States by Number of Customer
    visualize_customer_distribution(customer_locations)
    state_counts = state_level_analysis(customer_locations)

    print("Top 5 States by Customer Count:")
    print(state_counts.head())
```
